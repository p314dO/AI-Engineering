## Encontrar conectores clave
Es tu primer día de trabajo en DataSciencester, y el vicepresidente de redes está lleno de preguntas sobre tus usuarios. Hasta ahora no había tenido a quién preguntar, así que está muy contento de tenerte en el equipo.

En concreto, quiere que identifiques quiénes son los "conectores clave" entre los científicos de datos. Para ello, te proporciona un volcado de toda la red de DataSciencester. (En la práctica, no es habitual que te den los datos que necesitas. El capítulo 9 está dedicado a la obtención de datos).

¿Qué aspecto tiene este volcado de datos? Consiste en una lista de usuarios, cada uno representado por un `dict` (diccionario en Python) que contiene el `id`(que es un número) de ese usuario y `name`(que, en una de las grandes coincidencias cósmicas, rima con el del usuario id):

```python
users = [
    { "id": 0, "name": "Hero" },
    { "id": 1, "name": "Dunn" },
    { "id": 2, "name": "Sue" },
    { "id": 3, "name": "Chi" },
    { "id": 4, "name": "Thor" },
    { "id": 5, "name": "Clive" },
    { "id": 6, "name": "Hicks" },
    { "id": 7, "name": "Devin" },
    { "id": 8, "name": "Kate" },
    { "id": 9, "name": "Klein" }
]
```

También te proporciona los datos de “amistad”, representados como una lista de pares de identificadores:

```python
friendship_pairs = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
                    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

Por ejemplo, la tupla (0, 1)indica que el científico de datos con id0 (Hero) y el científico de datos con id1 (Dunn) son amigos. La red se ilustra en la Figura 1-1 .

 ![descripcion](./images/dsf2_0101.png)

Representar las amistades como una lista de pares no es la forma más sencilla de trabajar con ellas. Para encontrar todas las amistades del usuario 1, hay que recorrer cada par buscando aquellos que contengan el número 1. Si hubiera muchos pares, esto llevaría mucho tiempo.

En cambio, vamos a crear una tabla dictdonde las claves sean usuarios idy los valores sean listas de amigos id. (Buscar información en una tabla dictes muy rápido).

Todavía tendremos que examinar cada par para crearlo dict, pero solo tenemos que hacerlo una vez, y después obtendremos búsquedas económicas:

```python
#Initialize the dict with an empty list for each user id:
friendships = {user["id"]: [] for user in users}
friendships
```

```python
# And loop over the friendship pairs to populate it:
for i, j in friendship_pairs:
    friendships[i].append(j) # Add j as a friend of user i
    friendships[j].append(i) # Add i as a friend of user j

friendships
```

Ahora que tenemos las amistades en un `dict`, podemos hacer preguntas fácilmente a nuestro gráfico, como "¿Cuál es el número promedio de conexiones?"

Primero hallamos el número total de conexiones, sumando las longitudes de todas las `friends` listas:

```python
def number_of_friends(user):
    """How many friends does _user_ have?"""
    user_id = user["id"]
    friends_ids = friendships[user_id]
    return len(friends_ids)

total_connections = sum(number_of_friends(user) for user in users)
total_connections
```

Y luego simplemente dividimos por el número de usuarios:

```python
num_users = len(users)                            # length of the users list
avg_connections = total_connections / num_users   # 24 / 10 == 2.4
avg_connections
```

También es fácil encontrar a las personas con más contactos: son aquellas que tienen mayor número de amigos.

Como no hay muchos usuarios, podemos simplemente ordenarlos de "más amigos" a "menos amigos":

```python
# Create a list(user_id, number_of_friends)
num_friends_by_id = [(user["id"], number_of_friends(user)) for user in users]

num_friends_by_id.sort(
    key=lambda id_and_friends: id_and_friends[1],
    reverse=True
)
```

```python
num_friends_by_id
```

Podemos entender lo que hemos hecho como una forma de identificar a las personas que son, de alguna manera, fundamentales para la red. De hecho, lo que acabamos de calcular es la métrica de centralidad de grado de la red ( Figura 1-2 ).

![descripcion](./images/dsf2_0102.png)

Esto tiene la ventaja de ser bastante fácil de calcular, pero no siempre da los resultados que uno desearía o esperaría. Por ejemplo, en la red DataSciencester, Thor(`id` 4) solo tiene dos conexiones, mientras que Dunn(`id`1) tiene tres. Sin embargo, al observar la red, intuitivamente parece que Thor debería ser más central. Por ejemplo, `Hero (id 0)` no tiene amigos en común con `Klein (id 9)`, pero comparten intereses en Java y big data.

Es fácil crear una función que encuentre usuarios con un interés determinado:

```python
def data_scientists_who_like(target_interest):
    """Find the ids of all users who like the target interest."""
    return [user_id
            for user_id, user_interest in interests
            if user_interest == target_interest
           ]
```

Esto funciona, pero tiene que examinar toda la lista de intereses para cada búsqueda. Si tenemos muchos usuarios e intereses (o si simplemente queremos realizar muchas búsquedas), probablemente sea mejor crear un índice que vaya desde los intereses hasta los usuarios.

```python
from collections import defaultdict

#Keys are interests, values are lists of user_ids with that interest
user_ids_by_interest = defaultdict(list)

for user_id, interest in interests:
    user_ids_by_interest[interest].append(user_id)
```

Y otro de los usuarios a los intereses:

```python
# Keys are user_ids, values are lists of interests for that user_id.
interests_by_user_id = defaultdict(list)

for user_id, interest in interests:
    interests_by_user_id[user_id].append(interest)
```

Ahora es fácil encontrar quién tiene más intereses en común con un usuario determinado:
- Iterar sobre los intereses del usuario.
- Para cada interés, itere sobre los demás usuarios que tengan ese mismo interés.
- Lleva la cuenta de cuántas veces nos vemos, usuario.
En código:

```python
def most_common_interests_with(user):
    return Counter(
        interested_user_id
        for interest in interests_by_user_id[user["id"]]
        for interested_user_id in user_ids_by_interest[interest]
        if interested_user_id != user["id"]
    )
```

## Salarios y experiencia
Justo antes de ir a almorzar, el vicepresidente de relaciones públicas te pregunta si puedes darle algunos datos curiosos sobre cuánto ganan los científicos de datos. Los datos salariales son, por supuesto, confidenciales, pero logra proporcionarte un conjunto de datos anónimo que contiene el salario de cada usuario `salary`(en dólares) y `tenure` su antigüedad como científico de datos (en años):

```python
salaries_and_tenures = [(83000, 8.7), (88000, 8.1),
                        (48000, 0.7), (76000, 6),
                        (69000, 6.5), (76000, 7.5),
                        (60000, 2.5), (83000, 10),
                        (48000, 1.9), (63000, 4.2)]
```

El primer paso lógico es graficar los datos (lo cual veremos cómo hacer en el Capítulo 3 ). Puedes ver los resultados en la Figura 1-3 .

![image](./images/dsf2_0103.png)

Parece claro que las personas con más experiencia tienden a ganar más. ¿Cómo puedes convertir esto en un dato curioso? Tu primera idea es observar el salario promedio para cada período de antigüedad:

```python
# Keys are years, values are lists of the salaries for each tenure.
salary_by_tenure = defaultdict(list)

for salary, tenure in salaries_and_tenures:
    salary_by_tenure[tenure].append(salary)

# Keys are years, each value is average salary for that tenure.
average_salary_by_tenure = {
    tenure: sum(salaries) / len(salaries)
    for tenure, salaries in salary_by_tenure.items()
}
average_salary_by_tenure
```

Esto resulta no ser particularmente útil, ya que ninguno de los usuarios tiene la misma antigüedad, lo que significa que solo estamos informando los salarios de los usuarios individuales:

```python
{0.7: 48000.0,
 1.9: 48000.0,
 2.5: 60000.0,
 4.2: 63000.0,
 6: 76000.0,
 6.5: 69000.0,
 7.5: 76000.0,
 8.1: 88000.0,
 8.7: 83000.0,
 10: 83000.0}
```

Podría ser más útil agrupar los períodos de servicio:

```python
def tenure_bucket(tenure):
    if tenure < 2:
        return "less than two"
    elif tenure < 5:
        return "between two and five"
    else:
        return "more than five"
```

Entonces podemos agrupar los salarios correspondientes a cada categoría:

```python
# Keys are tenure buckets, values are lists of salaries for that bucket.
salary_by_tenure_bucket = defaultdict(list)

for salary, tenure in salaries_and_tenures:
    bucket = tenure_bucket(tenure)
    salary_by_tenure_bucket[bucket].append(salary)
```

Y finalmente, calcule el salario promedio para cada grupo:

```python
# Keys are tenure buckets, values are average salary for that bucket.
average_salary_by_bucket = {
  tenure_bucket: sum(salaries) / len(salaries)
  for tenure_bucket, salaries in salary_by_tenure_bucket.items()
}
```

¿Qué es más interesante?

Una forma sencilla (aunque no especialmente emocionante) de encontrar los intereses más populares es contar las palabras:
1. Escriba cada interés en minúscula (ya que los diferentes usuarios pueden o no escribir sus intereses con mayúscula inicial).
2. Divídelo en palabras.
3. Cuenta los resultados.

En código:

```python
words_and_counts = Counter(
    word
    for user, interest in interests
    for word in interest.lower().split()
)
words_and_counts
```

Esto facilita la tarea de enumerar las palabras que aparecen más de una vez:

```python
for word, count in words_and_counts.most_common():
    if count > 1:
        print(word, count)
```
