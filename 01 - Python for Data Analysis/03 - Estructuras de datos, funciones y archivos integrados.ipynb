## 3.1 Estructuras de datos y secuencias
Las estructuras de datos de Python son sencillas pero potentes. Dominar su uso es fundamental para convertirse en un programador de Python competente. Comenzaremos con tuplas, listas y diccionarios, que son algunos de los tipos de secuencia más utilizados .

### Tupla
Una tupla es una longitud fija, Secuencia inmutable de objetos Python que, una vez asignados, no se pueden modificar. La forma más sencilla de crear una es con una secuencia de valores separados por comas y encerrados entre paréntesis:

```python
tup = (4,5,6)
tup
```

En muchos contextos, los paréntesis pueden omitirse, por lo que aquí también podríamos haber escrito:

```python
tup = 4,5,6
```

```python
tup
```

Puedes convertir cualquier secuencia o iterador en una tupla invocando tuple:

```python
tuple([4,0,2])
```

```python
tup = tuple('string')
```

```python
tup
```

Se puede acceder a los elementos con corchetes [ ] como con la mayoría de los demás tipos de secuencias. Al igual que en C, C++, Java y muchos otros lenguajes, las secuencias se indexan desde 0 en Python:

```python
tup[0]
```
